```python
import matplotlib.pyplot as plt

# Sample data: a small set of values
data=[5,7,7,8,7,9,10,10,10,11,12]

#create the histogram
plt.hist(data, bins=5,edgecolor='black')

#add labels and title
plt.xlabel('Numbers')
plt.ylabel('Count')
plt.title('Histogram of Data')

#show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(34) # For reproductibility
data = np.random.randint(15, 200, 15000) #Generate 15000 ages between 15 to 15000

#Create histogram
plt.hist(data, bins=30,edgecolor='black',color='skyblue')

# Set the title and labels for the plot
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
#box plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    'Semister': ['sem 1','sem 1','sem 1','sem 1','sem 2','sem 2','sem 3','sem 4'],
    'hours_studied':[5,8,6,4,11,12,8,10]
}

#create a dataFrame
df=pd.DataFrame(data)

#create the boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='Semister',y='hours_studied',data=df)

plt.title('Students performance: Hours')
plt.xlabel('Semister')
plt.ylabel('Hours')

#create a box plot
sns.boxplot(x='Semister',y='hours_studied',data=df)
```
