```python
# data frame 1
import pandas as pd
dict1 ={'customer_id':[1,2,3,4,5,6],'Product':['Oven','Telivision','Oven','Fridge','Ac','Fan']}

df1 = pd.DataFrame(dict1)
df1
```

```python
# data 2
dict2 ={'customer_id':[2,4,6,8],'State':['California','Florida','Texas','Ohio']}
df2 = pd.DataFrame(dict2)
df2
```

```python
# inner join in python pandas
inner_join_df=pd.merge(df1,df2,on='customer_id',how='inner')
inner_join_df
```

```python
# outer join in pandas
outer_join_df=pd.merge(df1,df2,on='customer_id',how='outer')
outer_join_df
```

```python
# left join in python pandas
left_join_dfs=pd.merge(df1,df2,on='customer_id',how='left')
left_join_dfs
```

```python
#right join

right_join_dfs=pd.merge(df1,df2,on='customer_id',how='right')
right_join_dfs
```

```python
# Data 1
import pandas as pd
data1={'customer_name':['john','alok','jorge'], 'Product':['oven','television','oven']}
da1 = pd.DataFrame(data1)
da1
```

```python
#Data 2
data2={'Prize':[200,300,400],'Product':['oven','television','oven']}
da2 = pd.DataFrame(data2)
da2
```

```python
#inner join for data 1&2
inner_join_df=pd.merge(da1,da2,on='Product',how='inner')
inner_join_df
```

```python
#outer join on data 1&2
outer_join_df=pd.merge(da1,da2,on='Product',how='outer')
outer_join_df
```

```python
# left join
left_join_dfs=pd.merge(da1,da2,on='Product',how='left')
left_join_dfs
```

```python
#right join
right_join_dfs=pd.merge(da1,da2,on='Product',how='right')
right_join_dfs
```
