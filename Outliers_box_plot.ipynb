```python
import numpy as np

# sample data
data = [10,12,14,13,100,87,5,15,16,14,13,11]

# step -1 Calculate the percentile
Q1=np.percentile(data, 25)
Q3 = np.percentile(data, 75)


IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Finding  the outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#create a sample dataset with some outliers
np.random.seed(10)
data = pd.DataFrame({
    'value' : np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10,1,10)])
})

#Calculate the IQR for box plot-based outlier detection
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

#Define a threshold for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]

print(f"Outliers based on box plot criteria:\n{outliers}")

plt.figure(figsize=(8,6))
sns.boxplot(x=data['value'])
plt.title('Box Plot')

# violin plot with box plot inside for comparison
plt.subplot(1,2,2)
sns.violinplot(x=data['value'])
plt.title('Violin Plot')
plt.tight_layout()
plt.show()
```
