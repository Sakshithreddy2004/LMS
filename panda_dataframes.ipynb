```python
#sample d
import pandas as pd

data = {
    'Transaction_ID': [1,2,3,4,5],
    'Product_ID': [101,102,103,101,104],
    'Product_Name': ['shirt','pants','shoes','Hat','shirt'],
    'Customer_ID': [201,202,203,201,204],
    'Quantity': [2,3,1,1,2],
    'Price_Per_Unit': [30,40,50,30,15],
}
customer_data = {
    'CustomerID': [101,102,103,104],
    'CustomerName': ['Alice','Bob','Charlie','David'],
    'Age': [30,35,40,25],
    'City': ['New York','Los Angeles','Chicago','Huston'],

}

#convert data into Dataframe
sales_df = pd.DataFrame(data)
customers_df = pd.DataFrame(customer_data)

print("Sales DataFrame:")
print(sales_df.head())

# 1. Exploring the dataset using shape and describe

print("\nShape of sales data:", sales_df.shape)
print("\nDescription of sales data:")
print(sales_df.describe())
print("Sales dataframe",sales_df)
print("\nCustomers dataframe:\n",customers_df)
```
